## 1. Topic

### Policy change: in the policy change analysis, we'll use the pre-post and difference-in-difference methods to assess the potential effects of the three following drug prescription policy changes on 1) the volume of opioids prescription and 2) drug overdose deaths. 

1. Florida enacted several measures in correspondence to the prescribing and abuse of pain killers, Effective February, 2010

2. Texas Medical Board adopted regulations on treating pain with controlled substances and published guidelines, Effective January 4, 2007

3. Washington State adopted a regulation on prescribing of opioids for pain treatment, Effective Jan 2, 2012.


## 2. Project Questions

### Pre-post approach questions

1. Are opioid shipments and overdose deaths right after the policy went into effect in February, 2010 in Florida different from those before the policy went into effect?  

2. Are overdose deaths right after the policy went into effect on January 4, 2007 in Texas different from those before the policy went into effect?  

3. Are opioid shipments and overdose deaths right after the policy went into effect on Jan 2, 2012 in Washington different from those before the policy went into effect?  

### Difference-in-difference approach questions

1. For Florida policy changing occured on Feb, 2010, does its policy change effectively reduce both opioid shipments and overdose deaths? 

2. For Texas policy changing occured on Jan, 2007, does its policy change effectively reduce overdose deaths? 

3. For Washington's policy changing occured on Jan, 2012, did its policy change effectively reduce both opioid shipments and overdose deaths? 


## 3. Project Hypothesis

### Pre-post approach hypotheses

1. Had the opioid policy not gone into effect, opioid shipments and overdose deaths for Florida after 2010 would have looked more or less like it looked before 2010.

2. Had the opioid policy not gone into effect, overdose deaths for Texas after 2007 would have looked more or less like it looked before 2007.

3. Had the opioid policy not gone into effect, opioid shipments and overdose deaths for Washington after 2012 would have looked more or less like it looked before 2012. 

### Difference-in-difference approach hypotheses

1. For Florida, our hypothesis is "The magnitude of decrease in both opioid shipments and overdose deaths of Florida is larger than other states (those states exhibit similar trends) in the same time.
