## 1. Topic

### Policy change: in the policy change analysis, we'll perform a pre-post analysis as well as a difference-in-difference analysis to assess the potential effects of the three following policy changes:

1. Florida enacted several measures in correspondence to the prescription and abuse of pain killers, effective February, 2010

2. Texas Medical Board adopted regulations on treating pain with controlled substances and published guidelines, effective January 4, 2007

3. Washington State adopted a regulation on prescription of opioids for pain treatment, effective Jan 2, 2012.


## 2. Project Questions

### Pre-post approach questions


1. Do we see a significant decrease in opioid shipments and overdose deaths after the policy is implemented in Florida in February, 2010?
<br>
2. Do we see a significant decrease in opioid shipments and overdose deaths after the policy is implemented in Texas in January, 2007?
<br>
3. Do we see a significant decrease in opioid shipments and overdose deaths after the policy is implemented in Washington in January, 2012?

### Difference-in-difference approach questions

1. In regards to the Florida policy change from on February, 2010, do we see a significant reduction in both opioid shipments and overdose deaths as compared to other states? 
<br>
2. For the Texas policy change from January, 2007, do we see a significant reduction in overdose deaths as compared to other states? 
<br><
3. For Washington's policy change from January, 2012, do we see a significant reduction both opioid shipments and overdose deaths as compared to other states?  


## 3. Project Hypothesis

### Pre-post approach hypotheses

1. Had the opioid policy not gone into effect, opioid shipments and overdose deaths for Florida after 2010 would have looked more or less like it looked before 2010.

2. Had the opioid policy not gone into effect, overdose deaths for Texas after 2007 would have looked more or less like it looked before 2007.

3. Had the opioid policy not gone into effect, opioid shipments and overdose deaths for Washington after 2012 would have looked more or less like it looked before 2012. 

### Difference-in-difference approach hypotheses

1. The magnitude of decrease in both opioid shipments and overdose deaths of Florida is larger than other states (those states exhibit similar trends) in the same time.

2. The magnitude of decrease in overdose deaths of Texas is larger than other states (those states exhibit similar trends) in the same time.

3. The magnitude of decrease in both opioid shipments and overdose deaths of Washington is larger than other states (those states exhibit similar trends) in the same time.
